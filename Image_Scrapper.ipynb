#####The Original Dataset doesn't have any images, now we want to download some images to represent the product, this will make visualization on the product recommendation easier.

```python
#Lukas updated 25 February 2022
sheet_url = 'https://drive.google.com/file/d/1Re8bNtiVZhutdNLpgEZmNm3EcLKoBMCY/view?usp=sharing'
sheet_url ='https://drive.google.com/uc?id=' + sheet_url.split('/')[-2]
Data_Ecommerce = pd.read_csv(sheet_url,encoding= 'unicode_escape')
Data_Ecommerce.head()
```

```python
from pathlib import Path
root = Path().cwd()/"images"

from jmd_imagescraper.core import * # dont't worry, it's designed to work with import *
```

```python
# You can edit the parameters, but for now, we only need one image per product.
params = {
    "max_results": 1,             # this can go up to 477 at the time of writing
    "img_size":    ImgSize.Cached, 
    "img_type":    ImgType.Photo,
    "img_layout":  ImgLayout.Square,
}

```

```python
import numpy as np
item_counts = Data_Ecommerce["Description"].value_counts()
item_counts = pd.DataFrame(item_counts)
item_counts = item_counts.sort_values(by=['Description'], ascending = False)
item_counts.head(100)
```

```python
item_counts.quantile(0.8)

df=item_counts
```

```python
df[df['Description'].between(150, 249)].index
```

```python
a = 1
df=item_counts
b = len(df[df['Description'].between(50, 149)])
# This means we download the images with 50 until 149 transactions. The product that doesnt have this many transactions is not included, 
# because if its too much, the google colab display would crash.
```

```python
#delete all images
rmtree(root)
```

```python
# I add buy words because this is for ecommerce and the ecommerce algorithm favors the word buy
# df['Description'].between(50, 149) means we download the images with 50 until 149 transactions. The product that doesnt have this many transactions is not included, 
# because if its too much, the google colab display would crash.
# you can edit it to df[df['Description'] >= 200 to only choose items with 200 or more transactions
a = 1
for i in df[df['Description'].between(50, 149)].index :
  i2 = str(i+" buy")
  duckduckgo_search(root, i2, i2, **params)
  print("image of",a,'from total of',b)
  a = a+1
```
